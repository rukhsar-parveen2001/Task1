```python
import pandas as pd
```

```python
# Sample DataFrame
```

```python
data = {
    'CustomerID': [1, 2, 3, 4],
    'Date_of_Birth': ['1990-01-01', None, '1985-05-10', '2000-12-31'],
    'Transaction_Date': ['01/02/2023', '02-03-2023', '2023-04-05', '2023/05/06'],
    'Amount': [100, 200, None, 400]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
# 1. Handle missing values
```

```python
df['Date_of_Birth'] = df['Date_of_Birth'].fillna('1900-01-01')
df['Amount'] = df['Amount'].fillna(df['Amount'].mean())
```

```python
print(df)
```
